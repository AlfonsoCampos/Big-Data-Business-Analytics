##Scraping básico con Python

En este notebook tienes que realizar un "scraping" básico, para obtener las URLs del BOE del día.

Para ello tendrás que aplicar una expresión XPATH que extraiga las URLs que nos interesan.

Para poder ejecutar este notebook correctamente, necesitas las dos siguientes librerías instaladas en tu entorno python:

    requests
    lxml
    
Para instalarlas, tienes que ejecutar lo siguiente:

```python
import requests
import lxml
```

Vamos a centranos en la sección "Disposiciones Generales" del BOE del día 31/7/2015

```python
response = requests.get('http://www.boe.es/boe/dias/2015/07/31/')
content = response.content
```

```python
print content
```

Una vez tenemos descargado el contenido de la página que nos interesa, lo que queremos es aplicar una expresión XPATH que nos de las URLs de todos los PDFs de la página.

Lo más recomendable es estudiar el código fuente HTML de los elementos que nos interesa extraer. Por ejemplo, si estás usando Chrome para visualizar la página, sitúate sobre uno de los enlaces PDF y selecciona "Inspect Element" para que se abra la consola donde se muestran los detalles del código fuente HTML del elemento seleccionado.

A continuación, mediante prueba y error, ve aplicando la expresión XPATH que crees que más se aproxima a lo que quieres extraer (recuerda que lo que queremos extraer es la URL que está en el atributo href de los enlaces).

La siguiente expresión puede ser un buen punto de partida:

```python
xpath_expr = '//li/a'
```

Para aplicar la expresión XPATH a nuestra página, y ver el resultado, ejecuta las siguientes líneas:

```python
from lxml import html
tree = html.fromstring(content)
results = tree.xpath(xpath_expr)
```
